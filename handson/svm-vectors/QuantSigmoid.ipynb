```python

import pandas
import numpy
```

```python

bits = 16
formats = pandas.DataFrame({
    'm': range(1, bits),
})
formats['n'] = bits - formats.m 
formats['name'] = 'Q'+formats.m.astype(str)+'.'+formats.n.astype(str)
formats['range_min'] = -numpy.power(2, (formats.m-1))
formats['range_max'] = numpy.power(2, (formats.m-1))
formats['resolution'] = numpy.power(2.0, -formats.n)
formats = formats.set_index('name')
formats     
```

```python

from scipy.special import expit

inputs = numpy.linspace(-8, 8, 256)
outs = expit(inputs)
df = pandas.DataFrame(dict(inp=inputs, out=outs))
df

# TODO: convert output to fixed-point Q7 format
# TODO: serialize as C 8bit array
# compare output to sigmoidTable_q7
```
